```python
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
load_dotenv(override=True)
csv_path = os.getenv('CSV_PATH')
csv_path
```

```python
df = pd.read_csv(csv_path)
df.describe()
```

df = df[(df['gpt_model'] == 'chatgpt-4o') & (df['target_character'] == 2) & (df['pe'] == 'got') & (df['seed'].isin([2, 3, 6, 11, 13]))]

```python
df.columns
```

```python
df = df[df['fewshot'] == False]
df = df[df['evaluator'] == 'hr']
df = df[df['feedback_type'] == 'default']
df = df[df['exp_name'] != 'vision']
df = df[df['Evaluation/llm_iteration'] <= 6]
df
```

```python
df.groupby(['gpt_model', 'pe', 'target_character']).agg({'Evaluation/acc_imp_perc': ['count']})
```

```python
base_df = df[(df['n_self_alignment'] == 0) & (df['Evaluation/llm_iteration'] == 1)]
base_df = base_df.groupby(['gpt_model', 'pe', 'target_character']).agg({'Evaluation/acc_imp_perc': ['mean']})
base_df.columns = ['Base']
base_df = base_df.reset_index()
base_df
```

```python
sa_df = df[(df['n_self_alignment'] == 5) & (df['Evaluation/llm_iteration'] == 1)]
sa_df = sa_df.groupby(['gpt_model', 'target_character', 'pe']).agg({'Evaluation/acc_imp_perc': ['mean']})
sa_df.columns = ['+ Self-Alignment']
sa_df = sa_df.reset_index()
sa_df
```

```python
fb_df = df[(df['n_self_alignment'] == 0) & (df['Evaluation/llm_iteration'] == 6)]
fb_df = fb_df.groupby(['gpt_model', 'target_character', 'pe']).agg({'Evaluation/acc_imp_perc': ['mean']})
fb_df.columns = ['+ Feedback']
fb_df = fb_df.reset_index()
fb_df
```

```python
safb_df = df[(df['n_self_alignment'] == 5) & (df['Evaluation/llm_iteration'] == 6)]
safb_df = safb_df.groupby(['gpt_model', 'target_character', 'pe']).agg({'Evaluation/acc_imp_perc': ['mean']})
safb_df.columns = ['+ Feedback & Self-Alignment']
safb_df = safb_df.reset_index()
safb_df
```

```python
pk = ['gpt_model', 'target_character', 'pe']

merged = pd.merge(base_df, sa_df, on=pk).merge(fb_df, on=pk).merge(safb_df, on=pk)
merged
```

```python
mean = merged.groupby('gpt_model').agg({'Base': ['mean'], '+ Self-Alignment': ['mean'], '+ Feedback': ['mean'], '+ Feedback & Self-Alignment': ['mean']})
mean.columns = ['Base', '+ Self-Alignment', '+ Feedback', '+ Feedback & Self-Alignment']
mean.insert(0, 'pe', 'mean')
mean.insert(0, 'target_character', 'mean')
mean = mean.reset_index()
mean
```

```python
result = pd.concat([mean, merged])
result
```

```python
result.to_csv('self-alignment-and-feedback.csv', index=False)
```

```python

target_character_order = [1, 2, "mean"]  # 예: 특정 시나리오 번호와 "mean"
pe_order = ["cot", "tot", "got", "mean"]  # 예: Prompt Engineering 방식

result['target_character'] = pd.Categorical(result['target_character'], categories=target_character_order, ordered=True)
result['pe'] = pd.Categorical(result['pe'], categories=pe_order, ordered=True)


pivot_df = result.pivot_table(index=['target_character', 'pe'], columns='gpt_model', values=['Base', '+ Self-Alignment', '+ Feedback', '+ Feedback & Self-Alignment'])
pivot_df
```

```python
pivot_df.columns = pivot_df.columns.swaplevel(0, 1)
pivot_df.columns
```

```python
# 정렬 (주어진 Categorical 순서가 반영됨)
pivot_df = pivot_df.sort_index(level=['target_character', 'pe'], sort_remaining=False)

column_order = [
            (    'gpt-4o',                        'Base'),
            (    'gpt-4o',            '+ Self-Alignment'),
            (    'gpt-4o',                  '+ Feedback'),
            (    'gpt-4o', '+ Feedback & Self-Alignment'),
            (    'chatgpt-4o',                    'Base'),
            (    'chatgpt-4o',            '+ Self-Alignment'),
            (    'chatgpt-4o',                  '+ Feedback'),
            (    'chatgpt-4o', '+ Feedback & Self-Alignment')
]

# 결과 확인
pivot_df = pivot_df[column_order]
pivot_df
```

```python
def insert_str(src_str, idx, insert_str):
    # LaTeX 문자열을 줄 단위로 나눔
    lines = src_str.splitlines()
    
    # 지정된 줄 번호 뒤에 \midrule 삽입
    lines.insert(idx, insert_str)
    
    # 다시 LaTeX 문자열로 결합
    return "\n".join(lines)
```

```python
from IPython.display import display, Latex

# Generate the LaTeX string from the dataframe
tex_str = pivot_df.to_latex(
    buf=None, 
    escape=True, 
    bold_rows=False, 
    position='!t', 
    float_format="%.3f", 
    multicolumn_format='c',
    caption='Ablation study results for framework architecture, comparing accuracy across prompt engineering methods and scenarios.', 
    label='tab:result_pe'
)

# Perform replacements to modify the LaTeX string
replacements = {
    'target_character': 'Scenario',
    '& pe &': '& PE &',
    'gpt_model': 'LLM ($\\mathcal{M}$)',
    'table': 'table*',
    'Self-Alignment': 'SA',
    'Feedback': 'FB',
    '\\cline{1-10}\n': '',
    'cot': 'CoT',
    'tot': 'ToT',
    'got': 'GoT',
    'mean & mean': '\\multicolumn{2}{c|}{\\textbf{Mean}}',
    '\\multicolumn{4}{c}{gpt-4o}': '\\multicolumn{4}{c}{\\texttt{gpt-4o-2024-08-06}}',
    'chatgpt-4o': '\\texttt{chatgpt-4o-latest} (Dec 2024)',
    '\\begin{tabular}{llrrrrrrrr}': '\\begin{tabular}{ll|rrrr|rrrr}',
    '{c}{\\texttt{gpt-4o-2024-08-06}': '{c|}{\\texttt{gpt-4o-2024-08-06}'
}

for old, new in replacements.items():
    tex_str = tex_str.replace(old, new)

# Update the header to bold specific columns
original_header = ' &  & Base & + SA & + FB & + FB \\\\& SA & Base & + SA & + FB & + FB \\\\& SA \\\\'
modified_header = original_header.replace('Base', '\\textbf{Base}')
modified_header = modified_header.replace('+ SA', '\\textbf{+SA}')
modified_header = modified_header.replace('+ FB \\\\& SA', '\\textbf{+FB, SA}')
modified_header = modified_header.replace('+ FB', '\\textbf{+FB}')
tex_str = tex_str.replace(original_header, modified_header)

# Update the tabular header to adjust column widths
tabular_header = "\\begin{tabular}{p{1.5cm}p{1.5cm}|p{1.3cm}p{1.3cm}p{1.3cm}p{1.3cm}|p{1.3cm}p{1.3cm}p{1.3cm}p{1.3cm}}"
tex_str = tex_str.replace("\\begin{tabular}{ll|rrrr|rrrr}", tabular_header)

offset = 0
tex_str = insert_str(tex_str, 12 + offset, "\\midrule")
offset += 1
tex_str = insert_str(tex_str, 15 + offset, "\\midrule")

# Save the modified LaTeX string to a file
with open('ablation_study_detail.tex', 'w') as f:
    f.write(tex_str)

# Display the final LaTeX string
display(Latex(tex_str))
```
