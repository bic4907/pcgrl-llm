```python
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
```

```python
load_dotenv(override=True)
csv_path = os.getenv('CSV_PATH')
csv_path
```

```python
df = pd.read_csv(csv_path)
df.describe()
```

```python
df.columns
```

```python
df = df[df['fewshot'] == False]
df = df[df['feedback_type'] == 'default']
df = df[df['gpt_model'] == 'gpt-4o']
df = df[df['pe'].isin(['tot', 'got'])]
df
```

```python
df.groupby(['gpt_model', 'pe', 'evaluator']).agg({'Evaluation/acc_imp_perc': ['count']})
```

```python
base_df = df[(df['Evaluation/llm_iteration'] == 1)]
base_df = base_df.groupby(['gpt_model', 'pe', 'evaluator']).agg({'Evaluation/acc_imp_perc': ['mean']})
base_df.columns = ['Base']
base_df = base_df.reset_index()
base_df
```

```python
fb_df = df[(df['Evaluation/llm_iteration'] == 6)]
fb_df = fb_df.groupby(['gpt_model', 'pe', 'evaluator']).agg({'Evaluation/acc_imp_perc': ['mean']})
fb_df.columns = ['+ Feedback']
fb_df = fb_df.reset_index()
fb_df
```

```python
pk = ['gpt_model', 'pe', 'evaluator']

result = pd.merge(base_df, fb_df, on=pk)
result['Delta'] = result['+ Feedback'] - result['Base']

# Add mean row
mean_row = result.groupby(['gpt_model', 'evaluator']).agg({'Base': 'mean', '+ Feedback': 'mean', 'Delta': 'mean'}).reset_index()
# concat
mean_row['pe'] = ('mean')
result = pd.concat([result, mean_row], ignore_index=True)
result
```

```python

pe_order = ["cot", "tot", "got", "mean"]  # 예: Prompt Engineering 방식

result['pe'] = pd.Categorical(result['pe'], categories=pe_order, ordered=True)


pivot_df = result.pivot_table(index=['pe'], columns='evaluator', values=['Base', '+ Feedback', 'Delta'])
pivot_df
```

```python
pivot_df.columns = pivot_df.columns.swaplevel(0, 1)
pivot_df.columns
```

```python
# 정렬 (주어진 Categorical 순서가 반영됨)
pivot_df = pivot_df.sort_index(level=['pe'], sort_remaining=False)

column_order = [
            (    'hr',                        'Base'),
            (    'hr',            '+ Feedback'),
            (    'hr',            'Delta'),
            (    'llm',                  'Base'),
            (    'llm', '+ Feedback'),
            (    'llm', 'Delta'),
]

# 결과 확인
pivot_df = pivot_df[column_order]
pivot_df
```

```python
# change ('hr', 'Delta') column values to string with lambda
# pivot_df[('hr', 'Delta')] = pivot_df[('hr', 'Delta')].apply(lambda x: f'{x:.3f}')
pivot_df[('hr', 'Delta')] = pivot_df[('hr', 'Delta')].apply(
    lambda x: f'+{x:.3f}' if x > 0 else f'{x:.3f}'
)
pivot_df[('llm', 'Delta')] = pivot_df[('llm', 'Delta')].apply(
    lambda x: f'+{x:.3f}' if x > 0 else f'{x:.3f}'
)
pivot_df
```

```python
def insert_str(src_str, idx, insert_str):
    # LaTeX 문자열을 줄 단위로 나눔
    lines = src_str.splitlines()
    
    # 지정된 줄 번호 뒤에 \midrule 삽입
    lines.insert(idx, insert_str)
    
    # 다시 LaTeX 문자열로 결합
    return "\n".join(lines)
```

```python
from IPython.display import display, Latex


tex_str = pivot_df.to_latex(buf=None, escape=True, bold_rows=False, position='!h', float_format="%.3f", multicolumn_format='c',
                            caption='Ablation study results for framework architecture, comparing accuracy across prompt engineering methods and scenarios.', label='tab:self_evaluation')

tex_str = tex_str.replace('target_character', 'Scenario')
tex_str = tex_str.replace('pe &', 'PE &')
tex_str = tex_str.replace('evaluator &', '\\textbf{Evaluator} &')
tex_str = tex_str.replace('gpt_model', '')
# tex_str = tex_str.replace('table', 'table*')
# tex_str = tex_str.replace('Self-Alignment', 'SA')
# tex_str = tex_str.replace('Self-Alignment', 'SA')
# tex_str = tex_str.replace('Feedback', 'FB')
tex_str = tex_str.replace('\cline{1-10}\n', '')
tex_str = tex_str.replace('+ Feedback', '+FB')
tex_str = tex_str.replace('cot', 'CoT').replace('tot', 'ToT').replace('got', 'GoT')
tex_str = tex_str.replace('mean & mean', '\\multicolumn{2}{c|}{\\textbf{Mean}}')
tex_str = tex_str.replace('\multicolumn{3}{c}{hr}', '\multicolumn{3}{c|}{\\textbf{Heuristic (Oracle)}}')
tex_str = tex_str.replace('\multicolumn{3}{c}{llm}', '\multicolumn{3}{c}{\\textbf{LLM (Self-evaluate)}}')
tex_str = tex_str.replace('Delta', '$\\Delta{}$')
tex_str = tex_str.replace('mean &', '\\textbf{Mean} &')

original_header = ' &  & Base & + SA & + FB & + FB \\& SA & Base & + SA & + FB & + FB \\& SA \\\\'
modified_header = original_header.replace(
    'Base', '\\textbf{Base}'
).replace(
    '+ SA', '\\textbf{+SA}'
).replace(
    '+ FB \\& SA', '\\textbf{+FB, SA}'
).replace(
    '+ FB', '\\textbf{+FB}'
)

# 기존 문자열에서 헤더 교체
tex_str = tex_str.replace(original_header, modified_header)

tex_str = tex_str.replace('\\begin{tabular}{llrrrrrrrr}', '\\begin{tabular}{ll|rrrr|rrrr}')

tex_str = tex_str.replace('chatgpt-4o', '\\texttt{chatgpt-4o-latest} (Dec 2024)')

offset = 0
tex_str = insert_str(tex_str, offset + 11, "\\midrule")
offset += 1


tabular_header = "\\begin{tabular}{p{1.5cm}|p{0.73cm}p{0.73cm}p{0.73cm}|p{0.73cm}p{0.73cm}p{0.73cm}}"
# 기존 header를 새 header로 교체
tex_str = tex_str.replace("\\begin{tabular}{lrrlrrl}", tabular_header)

with open('self_evaluation.tex', 'w') as f:
    f.write(tex_str)
display(Latex(tex_str))
```
