```python
from zipfile import error

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('regular.csv')
df.describe()
```

```python
df = df[df['Evaluation/llm_iteration'].isin([1, 6])]
df.groupby(['pe', 'n_self_alignment', 'target_character', 'Evaluation/llm_iteration'])['Evaluation/acc_imp_perc'].count()
```

```python
base_df = df[(df['n_self_alignment'] == 0) & (df['Evaluation/llm_iteration'] == 1)]
base_df = base_df.groupby(['pe', 'target_character']).agg({'Evaluation/acc_imp_perc': ['mean']})
base_df.columns = ['Base']
base_df
```

```python
sa_df = df[(df['n_self_alignment'] == 5) & (df['Evaluation/llm_iteration'] == 1)]
sa_df = sa_df.groupby(['pe', 'target_character']).agg({'Evaluation/acc_imp_perc': ['mean']})
sa_df.columns = ['+ Self-Alignment']
sa_df
```

```python
fb_df = df[(df['n_self_alignment'] == 0) & (df['Evaluation/llm_iteration'] == 6)]
fb_df = fb_df.groupby(['pe', 'target_character']).agg({'Evaluation/acc_imp_perc': ['mean']})
fb_df.columns = ['+ Feedback']
fb_df
```

```python
safb_df = df[(df['n_self_alignment'] == 5) & (df['Evaluation/llm_iteration'] == 6)]
safb_df = safb_df.groupby(['pe', 'target_character']).agg({'Evaluation/acc_imp_perc': ['mean']})
safb_df.columns = ['+ Feedback & Self-Alignment']
safb_df
```

```python
merged = pd.merge(base_df, sa_df, on=['pe', 'target_character']).merge(fb_df, on=['pe', 'target_character']).merge(safb_df, on=['pe', 'target_character'])
merged = merged.reset_index()
# osrt PE with [cot, tot, got]

merged.columns = ['PE', 'Scenario', 'Initial Generation', '+ Self-Alignment', '+ Feedback', '+ Feedback & Self-Alignment']
merged
```
